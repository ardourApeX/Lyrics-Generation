# Train Markov Chain

```python
def GenerateTable(data, k = 4):
    transition_table = {}
    for i in range(len(data) - k):
        
        x = data[i : i + k] # Some text of  fixed length
        y = data[i + k] # Which word comes after text stored in x
        
        if transition_table.get(x) is None:
            transition_table[x] = {}
            transition_table[x][y] = 1
        else:
            if transition_table[x].get(y) is None:
                transition_table[x][y] = 1
            else:
                transition_table[x][y] += 1
    return transition_table
```

```python
def ConvertFreq2Probab(transition_table):
    '''This function is calculate frequency based upon frequency'''
    probability_table = transition_table.copy()
    
    for key in transition_table.keys():
        total = float(sum(transition_table[key].values()))
        
        for inner_key in transition_table[key].keys():
            probability_table[key][inner_key] = transition_table[key][inner_key]/total
    
    return probability_table
```

```python
def TrainMarkovChain(data, k = 4):
    
    TT = GenerateTable(data, k)
    PT = ConvertFreq2Probab(TT)
    
    return PT
```

```python
ProbabTable = TrainMarkovChain(text)
```

# Generate Text

```python
import numpy as np
```

```python
def GenerateText(query_text, ProbabTable, k = 4):
    roi = query_text[-k:]
    
    if ProbabTable.get(roi) is None:
        return " "
    else:
        possible_char = list(ProbabTable[roi].keys())
        possible_values = list(ProbabTable[roi].values())
        
        return np.random.choice(possible_char, p = possible_values)        
```

```python
GenerateText("apna ", ProbabTable)
```

```python
def ContinuousTextGeneration(starting_message, ProbabTable, max_len = 2000, k =4):
    sentence = starting_message
    roi = starting_message[-k:]
    
    for i in range(max_len):
        next_prediction = GenerateText(roi, ProbabTable, k)
        sentence += next_prediction
        roi = sentence[-k:]
        
    return sentence
```

```python
np.random.seed(11)
text = ContinuousTextGeneration(starting_message = "apna", ProbabTable = ProbabTable, k = 4, max_len = 2000)
```
